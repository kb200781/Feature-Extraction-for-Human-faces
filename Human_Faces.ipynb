```python
import os
import face_recognition
import pandas as pd

def extract_facial_features(image_path):
    # Load the image with face_recognition
    image = face_recognition.load_image_file(image_path)

    # Find face locations in the image
    face_locations = face_recognition.face_locations(image)

    # If no face is found, return None
    if not face_locations:
        return None

    # Get the facial landmarks for the first face (assuming only one face in the image)
    face_landmarks_list = face_recognition.face_landmarks(image, face_locations=[face_locations[0]])

    # Return the facial landmarks (e.g., chin, left eye, right eye, etc.)
    return face_landmarks_list[0]

def process_dataset(dataset_path):
    image_files = [f for f in os.listdir(dataset_path) if f.endswith(".jpg")]

    features_data = []

    for image_file in image_files:
        image_path = os.path.join(dataset_path, image_file)
        facial_features = extract_facial_features(image_path)

        if facial_features:
            feature_row = {'Image': image_file}
            for feature_name, points in facial_features.items():
                for i, (x, y) in enumerate(points):
                    feature_row[f"{feature_name}_x{i+1}"] = x
                    feature_row[f"{feature_name}_y{i+1}"] = y
            features_data.append(feature_row)
        else:
            print(f"No face found in {image_file}.")

    return features_data

if __name__ == "__main__":
    dataset_path = "/content/male_faces"
    features_data = process_dataset(dataset_path)

    # Convert the features data to a DataFrame
    df = pd.DataFrame(features_data)

    # Save the DataFrame to a CSV file
    csv_file = "malefacial_features.csv"
    df.to_csv(csv_file, index=False)

    print(f"Facial features extracted and saved to {csv_file}.")
```
